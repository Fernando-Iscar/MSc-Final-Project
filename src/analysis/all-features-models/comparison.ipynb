```python
%matplotlib inline

import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import os
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, make_scorer
```

```python
os.getcwd()
```

```python
train_set = pd.read_csv('../../output/clean_train_df.csv')
test_set = pd.read_csv('../../output/clean_test_df.csv')
```

```python
countries = list(train_set.filter(like='Country').columns)
```

```python
top20countries = ['Country_USA', 'Country_Germany', 'Country_UK & Northern Ireland', 'Country_Canada', 'Country_France', 'Country_India',
       'Country_Spain', 'Country_Brazil', 'Country_Poland', 'Country_Australia', 'Country_Netherlands', 'Country_Italy',
       'Country_Switzerland', 'Country_Israel', 'Country_Sweden', 'Country_Austria', 'Country_Russian Federation',
       'Country_Mexico', 'Country_Denmark', 'Country_Belgium']

AI_ML_devs = ['Data scientist or machine learning specialist',
 'Engineer, data',
 'Data or business analyst',
 'Scientist']

baseline_features = AI_ML_devs + top20countries
```

```python
baseline_pred = train_set.groupby(baseline_features)['Annual_salary'].median().reset_index()

baseline_pred
```

```python
y_test = test_set['Annual_salary']
test_baseline = test_set.merge(baseline_pred, on=[baseline_features])
rmse_baseline = np.sqrt(mean_squared_error(y_test, test_baseline["Annual_salary_y"]))
r2_baseline = r2_score(y_test, test_baseline["Annual_salary_y"])
```

```python
rmse_baseline, r2_baseline
```

```python
# Asegurándonos de que las columnas que indican la profesión son booleanas (0 o 1)
train_set[AI_ML_devs] = train_set[AI_ML_devs].astype(bool)

# Calculando la mediana del salario anual para las profesiones en AI_ML_devs en el conjunto de entrenamiento
medians = train_set.loc[train_set[AI_ML_devs].any(axis=1), 'Annual_salary'].median()

# Creando una columna de predicciones baseline en el conjunto de prueba, usando la mediana calculada
test_set['baseline_prediction'] = np.where(test_set[AI_ML_devs].any(axis=1), medians, np.nan)

# Asegurándonos de que todas las filas tienen una predicción
assert not test_set['baseline_prediction'].isna().any()

# Calculando RMSE y R2 para las predicciones baseline
rmse_baseline = np.sqrt(mean_squared_error(test_set['Annual_salary'], test_set['baseline_prediction']))
r2_baseline = r2_score(test_set['Annual_salary'], test_set['baseline_prediction'])

rmse_baseline, r2_baseline
```
